Visualizations

Univariate (One Variable)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings

warnings.filterwarnings('ignore')
```

Histogram

```python
# Generate a dataset
hs_customer_ages = np.random.normal(loc=35, scale=10, size=100)

# Plot the histogram
sns.histplot(hs_customer_ages, bins=20, color='skyblue', kde=True)
plt.title('Customer Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Boxplot

```python
# Generate a dataset
bp_salaries = np.random.normal(loc=750000, scale=10000, size=100)

# Plot the boxplot
sns.boxplot(x=bp_salaries, color='skyblue')
plt.xlabel('Salary (USD)')
plt.show()
```

Density Plot

```python
# Generate a dataset
dp_test_score = np.random.normal(loc=60, scale=10, size=1000)

# Plot 
sns.kdeplot(dp_test_score, color='skyblue', fill=True)
plt.title('Test Scores')
plt.xlabel('Score')
plt.ylabel('Density')
plt.show()
```

Bar Chart

```python

# Generate
bc_categories = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
bc_data = np.random.randint(10, 100, size=len(bc_categories))

# Plot
sns.barplot(x=bc_categories, y=bc_data, palette='Blues_d')
plt.title('Monthly Sales Performance')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

Pie Chart

```python

# Generate a dataset
bc_categories = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
bc_data = np.random.randint(10, 100, size=len(bc_categories))

# Plot
plt.figure(figsize=(8, 8))
plt.pie(bc_data, labels=bc_categories, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Sales by Product')
plt.show()
```

# Bivariate Visualization

## Scatter Plot

```python
# Generate
sp_num_points = 100
sp_ages = np.random.randint(22, 62, size=sp_num_points)
sp_income = sp_ages * 1000 + np.random.randn(sp_num_points) * 10000

# Plot the scatter plot
plt.figure(figsize=(10,6))
sns.scatterplot(x=sp_ages, y=sp_income, hue=sp_ages, palette='viridis', size=sp_income, sizes=(20, 200))
plt.title('Income vs. Age')
plt.ylabel('Income')
plt.xlabel('Age')
plt.legend(title='Age')
plt.show()
```

# Line plot

```python
# Generate a dateset
months = np.arange(1, 13)
years = ['2021', '2022', '2023']
lp_data = {year: np.random.randint(50, 150, size=12) for year in years}

# Plit
plt.figure(figsize=(10, 6))

# Define a for loop
for year, sales in lp_data.items():
    sns.lineplot(x=months, y=sales, marker='o', label=year)
    
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(months)
plt.legend(title='Year')
plt.show()
```

Heatmap

```python
# Generate a dataset
num_stocks = 10
stocks = [f'Stock {chr(65+i)}' for i in range(num_stocks)]
stock_prices = np.random.rand(num_stocks, num_stocks)

# Create correlation matrix
corr_matrix = np.corrcoef(stock_prices)

# Plot
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', xticklabels=stocks, yticklabels=stocks)
plt.title('Stock Prices')
plt.show()
```
